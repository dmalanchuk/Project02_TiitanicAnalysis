```python
import matplotlib.pyplot as plt
import pandas as pd

# Читання даних
gs = pd.read_csv('../data_source/processed/vizual.csv')


fig, ax = plt.subplots(figsize=(10, 6))
grouped_data = gs.groupby(['grouped_age', 'sex', 'pclass'])['total_survived_people'].sum().unstack(level='pclass')
grouped_data.plot(kind='bar', ax=ax)

ax.set_ylabel('Total Survived People')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Створюємо прикладні дані

res = pd.read_csv('../data_source/processed/res2.csv')

np.random.seed(0)
x = res['pclass']
y = res['rate']
z = res['age_with_avg'] 

# Створюємо DataFrame
data = pd.DataFrame({'x': x, 'y': y, 'z': z})

# Побудова графіка з трьома змінними
plt.figure(figsize=(8, 6))
scatter = plt.scatter(data['x'], data['y'], c=data['z'], cmap='viridis', s=100)
plt.colorbar(scatter, label='вік')
plt.title('Кореляційний графік з трьома змінними (x, y, z)')
plt.xlabel('клас квитка')
plt.ylabel('шанс вижити')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Завантажуємо дані
res = pd.read_csv('../data_source/processed/res2.csv')

# Перевіряємо, чи стовпці існують
if 'age_with_avg' in res.columns and 'pclass' in res.columns and 'rate' in res.columns:
    x = res['pclass']
    y = res['rate']
    z = res['age_with_avg']  # Шанси вижити

    # Створюємо DataFrame
    data = pd.DataFrame({'age_with_avg': x, 'pclass': y, 'rate': z})

    # Побудова графіка з трьома змінними
    plt.figure(figsize=(8, 9))
    scatter = plt.scatter(data['age_with_avg'], data['pclass'], c=data['rate'], cmap='viridis', s=100)
    plt.colorbar(scatter, label='Шанси вижити')
    plt.title('Кореляційний графік: Вік, Клас квитка та Шанси вижити')
    plt.xlabel('Вік (заповнено середніми значеннями)')
    plt.ylabel('щанс вижити')
    plt.show()
else:
    print("Стовпці 'age_with_avg', 'pclass' або 'rate' не знайдено в CSV файлі.")
```
